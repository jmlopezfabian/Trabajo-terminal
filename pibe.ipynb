```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/pibe.csv", sep = ";")
cdmx = df[df["Concepto"] == "Ciudad de México"]
cdmx_transpose = cdmx.transpose()
# primera fila como encabezado
cdmx_transpose.columns = cdmx_transpose.iloc[0]
cdmx_transpose = cdmx_transpose.iloc[1:]
cdmx_transpose = cdmx_transpose.reset_index()
plt.figure(figsize=(15, 4))
cdmx_transpose["log"] = cdmx_transpose["Ciudad de México"].apply(lambda x: np.log(x))
plt.plot(cdmx_transpose["index"], cdmx_transpose["log"])
```
